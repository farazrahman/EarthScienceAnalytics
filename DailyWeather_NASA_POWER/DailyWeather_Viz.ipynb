```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from weather_fetch import temp_to_df
```

```python
wx_df = temp_to_df(json_file='POWER_Point_Daily_20230101_20230331_040d71N_074d01W_LST.json')
wx_df.head()
```

```python
#parse dates
wx_df['date'] = wx_df['date'].apply(lambda x: datetime.strptime(x, "%Y%m%d"))
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data= wx_df[['T2M_MAX', 'T2M_MIN', 'T2M']])
plt.title('Min-Max Temp variation from Jan-March2023')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data= wx_df[['PRECTOTCORR']])
plt.title('Precipitation variation in from Jan-March2023')
plt.xlabel('Date')
plt.ylabel('Precipitation Corrected in mm/day')
plt.grid()
```
