```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from weather_fetch import temp_to_df
```

```python
wx_df = temp_to_df(json_file='POWER_Point_Daily_20230201_20230228_040d71N_074d01W_LST.json')
wx_df.head()
```

```python
#parse dates
wx_df['date'] = wx_df['date'].apply(lambda x: datetime.strptime(x, "%Y%m%d"))
```

```python
sns.lineplot(data= wx_df[['T2M_MAX', 'T2M_MIN', 'T2M']])
plt.title('Min-Max Temp variation in the Month of Feb_2023')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid()
```
